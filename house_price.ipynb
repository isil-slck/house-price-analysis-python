The scatter plot shows a very strong linear relationship between Price and Area. This was to be expected as often RE companies price their property per square foot.
```python
ax = df['Gender'].value_counts().plot(kind ='barh')
for i, v in enumerate(df['Gender'].value_counts()):
    ax.text(v + 3, i + .05, str(v), color='blue', fontweight='bold')

right_side = ax.spines["right"]
right_side.set_visible(False)

top_side = ax.spines['top']
top_side.set_visible(False)
```

```python
ax = sns.countplot(x="Year of sale", hue="Purpose", data=df);

right_side = ax.spines["right"]
right_side.set_visible(False)

top_side = ax.spines['top']
top_side.set_visible(False)
```

```python
find_ratio = df.groupby(['Year of sale', 'Purpose']).size().unstack(fill_value=0)
find_ratio['ratio'] = find_ratio['Investment']/find_ratio['Home']
#change index to column
find_ratio['Year of sale'] = find_ratio.index
find_ratio
```

```python
fig = plt.figure()

ax = find_ratio.plot.scatter(x= 'Year of sale',y = 'ratio' ,fontsize=10, sort_columns=True)
# Major ticks every 20, minor ticks every 5
major_ticks = np.arange(0, 1.2,0.1)
minor_ticks = np.arange(0, 1,0.05)

# And a corresponding grid
ax.grid(which='both')


ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)
# Or if you want different settings for the grids:
ax.grid(which='minor', alpha=1)
ax.grid(which='major', alpha=0.5)

plt.show();

```

In 2005, a large proportion of houses were purchased compared to 2004. However, in 2006, the house was purchased again until 2004. Since 2006, the rate has increased linearly, but the rate in 2005 was not reached. Since there is not enough data for 2009 and 2010, we cannot interpret these years.

2005 te 2004 e göre büyük oranda ev alımı yapılmıştır.Buna ragmen 2006 da yenıden 2004 teki kadar ev alımı yapılmıştır.2006 itibaren oran linear olarak artıs göstermıştır fakat 2005 teki orana ulaşılamamıştır. 2009 ve 2010 için yeterli veri olamadığından bu yılları yorumlayamıyoruz.

```python
#create price per area

df['price_per_area'] = df['Price']/df['Area (ft.)']
```

```python
ax = sns.boxplot(data=df, x="Year of sale", y="price_per_area",showmeans=True,
            meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"9"});

right_side = ax.spines["right"]
right_side.set_visible(False)

top_side = ax.spines['top']
top_side.set_visible(False)
```

In 2005 price per area did not change significatly in contrast to 2004,however in 2006 the price per area increased even though the sales of properties decreased.

```python
ax = sns.barplot(data=df, x='Year of sale', y="price_per_area", hue = "Mortgage");

right_side = ax.spines["right"]
right_side.set_visible(False)

top_side = ax.spines['top']
top_side.set_visible(False)
```

In 2008, there is a significant difference between mortgage and direct sold in terms of price per area. It can be ignored for data in 2009 and 2010 due to absence of relevant data.
